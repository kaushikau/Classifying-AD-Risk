# Abstract

Alzheimer’s disease (AD) is a progressive brain disorder characterized by the damage of neurons needed for memory, language, and thinking. Genetic variations in the Apolipoprotein E (APOE) and phosphatidylinositol binding clathrin assembly protein (PICALM) genes interfere with the clearance of amyloid-β and increase the risk of developing AD. We analyzed a multimodal dataset from research conducted by Dzianok and Kublik (2024), which includes functional (fMRI and EEG), blood test, and genetic data. We examined APOE and PICALM gene variants as well as factors collected from blood samples, assessed fMRI data from the Sternberg memory task to detect blood flow in the prefrontal cortex, and analyzed EEG data from the same task to distinguish any patterns of power changes in frequency bands. Our overall goal was to investigate whether differences in neural activity patterns as measured by fMRI and EEG as well as factors found in blood potentially serve as early indicators of Alzheimer's disease and predict genetic risk factors for Alzheimer's disease, specifically variations in the APOE and PICALM genes.

# Research Question

Can differences in neural activity as measured by fMRI and EEG as well as characteristics of blood profiles between people with and without Alzheimer's disease risk genes serve as early indicators of AD and predict genetic risk factors?

# Background and Prior Work

Alzheimer's disease is a brain disorder and a type of dementia characterized by cell degeneration in the brain and cognitive impairment. The disorder results in decline of memory, language, and overall cognitive function, and is highly linked to genetic factors. 

The e4 allele of the gene APOE (APOE4) is associated with increased amyloid-beta plaque accumulation which can impair neuronal function and connectivity (Yang et al., 2023). In addition, APOE4 can lead to neuroinflammation and oxidative stress, which are key pathological features of AD (Yang et al., 2023). Furthermore, the gene PICALM plays a role in clathrin-mediated endocytosis, which is crucial for synaptic vesicle recycling and synaptic platicity (Ando et al., 2022). Variations in the PICALM gene can therefore affect synaptic function and plasticity, potentially leading to synaptic loss and dysfunction, which are early indicators of neurodegenerative processes. A study by Ponomareva et al. (2020) found the GG allele of the PICALM gene to be an AD risk variant, and even linked the variant to alpha rhythm differences in EEG.

Blood profiles are another avenue in which risk factors for AD have been analyzed. According to Chen et al. (2017), there are various factors of blood cells that are abnormal in patients with Alzheimer's disease. These include decreased levels of hemoglobin, lymphocytes, and platelet distribution width (PDW), while other factors such as mean corpuscular volume (MCV), red cell distribution width (RDW-CW), and mean platelet volume (MPV) have increased levels. High levels of cholesterol in blood may also be a risk factor for AD development (Dzianok and Kublik, 2024). Differences in blood factors, therefore, might possibly be an early identifier of AD risk.

Previous research has also connected discrepancies in functional neuroimaging data to AD. Electroencephalography (EEG) recordings reveal brain electrical activity, as well as rhythms in neural activity at various frequency bands (Jafari et al., 2020). A longitudinal EEG study conducted by Coben et al. (1985) for a time period of 2.5 years revealed that there are significant changes in EEG power in people with AD, which included increases in power for the delta and theta frequency bands and decreases in power for the alpha and beta bands. This slowing of faster oscillations combined with the enhancement of slower oscillations has been a repeated finding of AD research, and is a common marker of AD risk.

Functional MRI (fMRI), on the other hand, can reveal differences in blood flow and volume using the blood oxygen-level dependent (BOLD) signal. A fMRI study done by Sperling et al. (2003) found that when compared to control patients, patients with Alzheimer's showed less activation in the hippocampus, a structure known for its role in memory and learning, during memory tasks. In addition, another study observed reduced activity in various regions of the prefrontal cortex (PFC), indicating that AD risk could be characterized by changes in fMRI responses in certain brain regions (Li et al., 2009).

Blood test, EEG, and fMRI data all have factors that have been found to be correlated with Alzheimer's disease risk, and therefore the motivation of this study is to identify blood and functional risk factors that could predict AD risk. Analyzing these risk factors early can lead to better diagnosis, more targeted treatments, and ultimately a better prognosis for individuals at risk for developing AD. We will use EEG and fMRI data collected during the Sternberg memory task, which is used to examine working memory and short-term memory. During this task, participants are shown a list of items (letters, in this experiment) for 3 seconds, asked to hold them in their memory for a retention period of 2 seconds, and then probed with an item three times, with a short interval between each one. They then must indicate whether that target item was in the original list or not. (Dzianok and Kublik, 2024). Since deficits in memory, specifically working memory, are characteristic of AD, this task is relevant to our analysis. In this study, we aim to analyze blood sample, EEG, and fMRI data collected by Dzianok and Kublik (2024) in an attempt to classify participants with and without the AD risk alleles present in the APOE and PICALM genes. 

## Hypothesis

Individuals with the genetic variants known to be risk factors (APOE4 and GG) for AD may illustrate differences in fMRI, EEG and blood test results. It is commonly known that individuals with AD show decreased BOLD signals, slowing of high frequency EEG oscillations and enhancement of low frequency oscillations, and abnormal levels of cholesterol, hemoglobin, hematocrit, and lymphocytes. Thus, we aim to predict whether or not having fMRI, EEG and blood test risk factors are associated with having the risk alleles. If we use logistic regression on these fMRI, EEG, and blood test risk factors, we hypothesize that our model will accurately predict whether our subjects are at risk for AD, which is characterized by having the genetic risk alleles APOE4 and GG in the genes APOE and PICALM.

## EEG Data

The EEG data used in this project is from the same OpenNeuro experiment. Since we are trying to see if we can separate the brain activity of participants with and without genetic risk factors for AD, we decided to use EEG data collected from participants during a specific task: the Sternberg memory task.

First, we created a function, `preprocessing_eeg` to pre-process the EEG data of all of the subjects. This function first loads in the data, then sets the reference to an average of all of the channels. Then, we needed to filter out the power line artifact from the data. The reason for this is shown in the power spectral density plot below (for Sub-01), where it is clear that there are spikes in the data at 50Hz, 150Hz, etc. To combat this, we created a notch filter that reduces this artifact in the data.

![PSD for Sub-01](psd_sub01.png)

The function then gets the events from the annotations in the raw data file and creates epochs: windows of time around the events. It then finally returns these epochs as an Epoch object.

````python
def preprocessing_eeg(file_path):
    # read in EEG file
    raw = mne.io.read_raw(file_path, preload=True).crop(tmax=100)
    
    # set reference
    raw.set_eeg_reference("average", projection=False, verbose=False)
    
    # filter out power line artifact (50Hz for this data)
    power_line = (50, 150, 250, 350, 450)
    raw = raw.notch_filter(freqs=power_line, method="spectrum_fit", filter_length="10s")

    # get events and creating dictionary of different events (obtained from 'task-sternberg_events.json')
    events = mne.events_from_annotations(raw)
    event_dict = {"Response: not true": 1,
              "Response: true": 2,
              "Low-demanding condition trial": 3,
              "High-demanding condition trial": 4,
              "Retention": 11,
              "Probe": 12}

    # create epochs from events
    epochs = mne.Epochs(raw, events[0], event_id=event_dict, preload=True)
    return epochs
````

We want to conduct analysis in a specific region of interest (ROI) - the prefrontal cortex (PFC). This region is associated with executive function and memory processing - since we are analyzing the risk factors of Alzheimer's disease, we expect we may see a difference in this area. To do so, we need to create a mask of this region of interest so that it can be analyzed further. This code was also run locally in a separate script in order to apply it to the subject fMRI files that were loaded locally.

First, a binary mask is created using the Harvard-Oxford cortical atlas. This identifies the PFC.

````python
# getting current working directory
data_dir=os.getcwd()

# creating the binary mask
atlas = datasets.fetch_atlas_harvard_oxford('cort-maxprob-thr50-1mm', data_dir=data_dir)
atlas_filename = atlas.filename
atlas_img = nib.load(atlas_filename)
labels = atlas.labels
````

Next, the code below stores the areas of the prefrontal cortex in a list and generates a list of indicies for the elements in `pfc_indicies`.

````python
pfc_labels = ['Frontal Pole', 'Insular Cortex', 'Superior Frontal Gyrus', 'Middle Frontal Gyrus', 'Inferior Frontal Gyrus']

pfc_indices = [i for i, label in enumerate(labels) if label in pfc_labels]
````

Then the following code generates a PFC mask by combining the ROIs identified by `pfc_indices`. First, it creates an empty boolean mask, then combines the specified ROIs. Next, the mask data was converted to an integer format for compatibility, and finally the resulting mask was saved to `pfc_mask.nii.gz`.

````python
# creating an empty mask
pfc_mask_data = np.zeros(atlas_img.shape, dtype=bool)

# combining regions to create the PFC mask
for i in pfc_indices:
    pfc_mask_data = np.logical_or(pfc_mask_data, atlas_img.get_fdata() == i)

# converting the mask data to int32
pfc_mask_data = pfc_mask_data.astype(np.int32)

# saving the PFC mask as a Nifti image
pfc_mask_img = nib.Nifti1Image(pfc_mask_data, affine=atlas_img.affine)
nib.save(pfc_mask_img, 'pfc_mask.nii.gz')
````

To ensure compatibility between the PFC mask and the rest of the subject's fMRI data, we need to resample the mask to match the target affine and shape of the subject fMRI data. The code below, uses the first subject's fMRI data as a reference. We started off by assigning `example_fmri_img` to subject-01 in order to obtain its affine transformation matrix and spatial dimensions. The PFC mask `pfc_mask_img` was then resampled to align with the affine and shape of the example fMRI image. This ensures that the PFC mask and subject fMRI data were in the same spatial orientation and dimensions.

After resampling, the new PFC mask was saved to `resampled_pfc_mask.nii.gz` for later use. To ensure we masked the correct area, we decided to visualize the resampled PFC mask as shown in the diagram below the code.

To complete this process we used NiftiMasker with the resampled PFC mask. NiftiMasker is a class from the NiLearn package that simplifies the extraction of time series data from a ROI in fMRI data. By using NiftiMasker, we can facilitate processing the fMRI data and further analyze the prefrontal cortex. The masker is saved as `pfc_masker` to later apply to subject fMRI data locally.

```python
# loading an example fMRI image to get the target affine and shape
```

Now that we have the raw BOLD signals saved as files, we can now average and load the each of the signals. The following function ```load_and_avg_bold_signal``` loads each BOLD signal, computes the means, then saves it in a dictonary. Afterwards, each of the average BOLD signals is transformed into a Pandas array.

````python
# empty list and dictonary for saving the files and loaded data
bold_files = []
bold_signals_dict = {}

# subject IDs
subjects_to_remove = {6, 22, 28, 29, 30}
sternberg_fmri_subjects = [f"sub-{i:02}" for i in range(1, 46) if i not in subjects_to_remove]

# storing each of the files in a list
for i in sternberg_fmri_subjects:
    file_path = f"data/preprocessed_fmri/saved_{i}_task-sternberg_bold_signal.npy"
    bold_files.append(file_path)

def load_and_avg_bold_signal(file_path, subject_id):
    # loads and averages each BOLD signal from each subject
    bold_signal = np.load(file_path)
    mean_bold_signal = np.mean(bold_signal)
    bold_signals_dict[subject_id] = mean_bold_signal
    return mean_bold_signal

# using function load_and_avg_bold_signal to calculate the average BOLD signal for each fmri subject in the PFC
for subject_id in sternberg_fmri_subjects:
    file_path = f"data/preprocessed_fmri/saved_{subject_id}_task-sternberg_bold_signal.npy"
    load_and_avg_bold_signal(file_path, subject_id)
````

````python
# creating a pandas dataframe to store subject ID and mean bold signal
bold_signals_df = pd.DataFrame.from_dict(bold_signals_dict, orient='index', columns=['mean_bold_signal'])


# changing index into integer variable and naming it participant_id for easier joining later
fmri_ids = []
for i in bold_signals_df.index:
    fmri_ids.append(int(i[4:]))
bold_signals_df['participant_id'] = fmri_ids

# setting participant_id as index
bold_signals_df = bold_signals_df.set_index('participant_id').sort_index()

print('Bold Signals Shape: ', bold_signals_df.shape)
bold_signals_df.head()
````

# Data Analysis and Results

## Determining Participants' Risk for AD From AD Risk Genes

First, we needed to separate the participants of the study based on whether they had the risk alleles for the two genes, APOE and PICALM, or not. According to Dzianok and Kublik (2024), the `e4` allele for APOE has previously been reported to be linked to an increased risk to developing AD. Furthermore, according to Ponomareva et al. (2020), the `G/G` variant for PICALM has been shown to be a risk factor for AD.

## Prefrontal Channel Power of EEG Frequency Bands

We decided to use the two prefrontal channels for our analysis, because previous research shows that prefrontal EEG has been associated with the predementia stages of Alzheimer's disease. Therefore, we analyzed frequency band power for `Fp1` and `Fp2`.

First, we created a dictionary that contains the frequency ranges for the five bands: `delta`, `theta`, `alpha`, `beta`, and `gamma`. Then, we created power dictionaries for each channel, with a separate list for each frequency band.

For each subject and for each frequency band, we calculated the power spectral density (PSD) using the Welch's method. After the PSDs were calculated, we got the average power across all of the epochs. For each frequency range, the function that computes PSDs uses multiple frequencies to calculate power, so we then found the max power for each frequency range, assigned to the variables `fp1_avg_power` and `fp2_avg_power`. Each of the max powers were then added to the power dictionaries created at the beginning, which were then converted to a data frame for easier analysis. The index was set to the `participant_id`, and all of the powers were converted from $V^2$ to $uV^2$, as the `compute_psd()` function returns values in volts, but EEG is measured in microvolts.

````python
%%capture
# creating dictionary of the frequency bands
frequency_bands = {'delta':(0,4), 'theta':(4,8), 'alpha':(8,12), 'beta':(12,30), 'gamma':(30,45)}

# creating a dictionary for the power values for each band for the two pre-frontal channels
fp1_power_dict = {'delta':[], 'theta':[], 'alpha':[], 'beta':[], 'gamma':[]}
fp2_power_dict = {'delta':[], 'theta':[], 'alpha':[], 'beta':[], 'gamma':[]}

# for each band in each subject, for each channel, calculate the psd; then get mean powers across epochs and get the highest power in the frequency range
for subject in epoched_data:
    for band in frequency_bands:
        # Fp1
        fp1_psds, fp1_freqs = subject.compute_psd(method='welch', fmin=frequency_bands[band][0], fmax=frequency_bands[band][1], picks=['Fp1']).get_data(return_freqs=True)
        fp1_avg_power = fp1_psds.mean(axis=0).max()
        fp1_power_dict[band].append(fp1_avg_power)
        
        # Fp2
        fp2_psds, fp2_freqs = subject.compute_psd(method='welch', fmin=frequency_bands[band][0], fmax=frequency_bands[band][1], picks=['Fp2']).get_data(return_freqs=True)
        fp2_avg_power = fp2_psds.mean(axis=0).max()
        fp2_power_dict[band].append(fp2_avg_power)

fp1_power_df = pd.DataFrame(fp1_power_dict).set_index(participants_genes.index)
fp2_power_df = pd.DataFrame(fp2_power_dict).set_index(participants_genes.index)

# compute_psd() units are V^2, so converting to uV^2 (EEG measured in microvolts)
fp1_power_df = fp1_power_df * (10**12)
fp2_power_df = fp2_power_df * (10**12)
````

Next, we renamed all of the columns to include the channel name, so that when all of these values are combined into one data frame, it is clear which channel each value is from.

````python
fp1_power_df = fp1_power_df.rename(columns={'delta':'fp1_delta', 'theta':'fp1_theta', 'alpha':'fp1_alpha',
                            'beta':'fp1_beta', 'gamma':'fp1_gamma'})
fp2_power_df = fp2_power_df.rename(columns={'delta':'fp2_delta', 'theta':'fp2_theta', 'alpha':'fp2_alpha',
                            'beta':'fp2_beta', 'gamma':'fp2_gamma'})
````

The two channels' data frames were then combined into one data frame that contains all of the EEG data, so that it is easier for our final analysis of classifying risk by using factors from all three of our datasets.

````python
EEG_data = fp1_power_df.join(fp2_power_df)
EEG_data.head()
````

## Classifying AD Risk Using Blood Test, EEG, and fMRI Data

Now that we have the factors from all three modalities: blood sample, EEG, and fMRI, our next step was the create a data frame, `all_data`, that contains all of the factors, as well as our binary variable, `Risk`, in order to use the factors to predict AD risk. Therefore, we joined all three of the data frames using their common index, `participant_id`.

````python
all_data = selected_blood_features.join(EEG_data)
all_data = all_data.join(bold_signals_df)
all_data['Risk'] = participants_genes['Risk']

print('All Data Shape: ', all_data.shape)
all_data.head()
````

The data frame above now contains blood, EEG, and fMRI factors for all 40 subjects for a total of 16 different factors (5 from blood test, 10 from EEG, and 1 from fMRI), as well as the `Risk` column containing the binary variable indicating whether the subject has the risk alleles for AD or not.

Since our final aim was to use a combination of multimodal data to classify AD risk, we now ran logistic regression on our data. We split the data using a 75/25 split, resulting in 30 participants in the training set and 10 participants in the testing set.

````python
# defining X and y for our data
X = all_data.iloc[:, :-1].values
y = all_data.iloc[:, -1].values

# splitting dataset into training and testing sets and standardizing values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=6)
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

# fitting a logistic regression model on the data
final_model = LogisticRegression()
final_model.fit(X_train, y_train)
````

We then created a data frame for the training data, and predicted the probability of risk using our model.

````python
df_train = pd.DataFrame(data=X_train, columns=all_data.columns[:-1])
df_train['Risk'] = y_train
train_probabilities = final_model.predict_proba(X_train)[:, 1]
y_pred_train = final_model.predict(X_train)
df_train['Probability of Risk'] = train_probabilities
df_train['Prediction of Risk'] = y_pred_train
df_train.head()
````

The data frame above now contains the predicted probability of risk for all of the subjects in our training dataset, as well as their predicted and actual `Risk` values. We then wanted to visualize how well the model predicted risk, so we plotted the confusion matrix for the training dataset.

````python
# creating confusion matrix
conf_matrix_train = metrics.confusion_matrix(y_train, y_pred_train)

# plotting confusion matrix
categories = [0,1]
fig, ax = plt.subplots()
ticks = np.arange(len(categories))
plt.xticks(ticks, categories)
plt.yticks(ticks, categories)
cbar_ticks_train = np.arange(0, conf_matrix_train.max()+1, 1)
sns.heatmap(pd.DataFrame(conf_matrix_train), cmap='Blues', cbar_kws={'label': 'Number of Participants', 'ticks': cbar_ticks_train});
plt.tight_layout()
plt.title('Confusion Matrix for Training Data');
plt.xlabel('Predicted Risk');
plt.ylabel('Actual Risk');
````

````python
print('Accuracy of Training Data Predictions: ', metrics.accuracy_score(y_train, y_pred_train))
````

Based on the confusion matrix plotted above, we can see that the model was fairly accurate for the training dataset, as the major diagonal has darker shades compared to the minor diagonal. The accuracy score printed above reveals the same thing: the model accurately predicts risk at about 80% accuracy for the training data.

Now that we've trained the model on the training data, we then used our testing data to see how accurate the model is.

````python
df_test = pd.DataFrame(data=X_test, columns=all_data.columns[:-1])
df_test['Risk'] = y_test

test_probabilities = final_model.predict_proba(X_test)[:, 1]
y_pred_test = final_model.predict(X_test)
df_test['Probability of Risk'] = test_probabilities
df_test['Prediction of Risk'] = y_pred_test
df_test.head()
````

Similar to the training set, `df_test` now contains the predicted and actual values of `Risk` and the probability of risk for each of the subjects. We now wanted to examine how well the model performed, so we plotted the confusion matrix for the testing data.

````python
# creating confusion matrix
conf_matrix_test = metrics.confusion_matrix(y_test, y_pred_test)

# plotting confusion matrix
categories = [0,1]
fig, ax = plt.subplots()
ticks = np.arange(len(categories))
plt.xticks(ticks, categories)
plt.yticks(ticks, categories)
cbar_ticks = np.arange(0, conf_matrix_test.max()+1, 1)
sns.heatmap(pd.DataFrame(conf_matrix_test), cmap='Blues', cbar_kws={'label': 'Number of Participants', 'ticks': cbar_ticks});
plt.tight_layout()
plt.title('Confusion Matrix for Testing Data');
plt.xlabel('Predicted Risk');
plt.ylabel('Actual Risk');
````

````python
print('Accuracy of Testing Data Predictions: ', metrics.accuracy_score(y_test, y_pred_test))
````

The confusion matrix plotted for the testing data above reveals that the model tends to predict no risk even when the subject actually does have the risk alleles. However, the accuracy score of the testing data reveals that the model's accuracy is at 70%, which is lower than it was for the training data, but is still relatively high.

# Discussion

Our study focused on the prediction of whether genetic variants known to be risk factors for Alzheimer's disease (AD), specifically the APOE4 and GG alleles in the APOE and PICALM genes, were present in individuals based on specific patterns in the three modalities of data that were analyzed: blood test, EEG, and fMRI.

The results indicate that there exist some differences in neural activity as well as blood biomarkers between individuals with the genetic risk alleles for AD. In terms of blood biomarkers, the most meaningful features that impacted the model predicting risk with only blood test factors were hematocrit, red cell distribution width (RDW-CW), mean corpuscular volume (MCV), percent of eosinophils, and low-density lipoprotein (LDL) cholesterol. We hypothesized that the blood factors that would play the largest role would be levels of cholesterol, hemoglobin, hematocrit, and lymphocytes. Therefore, although some of the factors that we predicted would have a large role in the model were accurate—cholesterol and hematocrit—we were not correct about all of them. MCV, RDW-CW, and percent of eosinophils have all been found to have correlations with AD risk, however, as MCV and RDW-CW have been shown to have decreased levels in AD patients (Chen et al., 2017), while eosinophil levels have been found to be slightly higher in individuals with AD risk (Dzianok and Kublik, 2023).

For differences in EEG signals between participants who had the risk alleles and did not, the topographic maps do not fully confirm whether the power of slow frequencies increase and the power of fast frequencies decrease for individuals with the AD risk alleles, which is a common marker for AD in EEG recordings (Coben et al., 1985). Furthermore, since we concentrated on isolating powers in the prefrontal region of the brain just to use as factors for our final model, our analysis did not reveal whether there is a meaningful difference between the two groups. However, the powers for all five frequency bands—delta, theta, alpha, beta, and gamma—for both prefrontal channels, Fp1 and Fp2, were used in our final model, and as our logistic regression model to predict risk performed relatively well, we note that although we cannot confirm that there is a significant difference in EEG signals between individuals with and without the AD risk alleles, they do seem to have some role as early indicators of AD risk.

Graphing the BOLD signals in the prefrontal cortex of 5 subjects that have the risk allele and 5 subjects who weren't at risk did not make it clear whether or not there is a trend between BOLD intensity and AD risk. Previous research shows that AD patients have disruptions of neural networks associated with memory processing (Vemuri, P., et al., 2012); furthermore, Dzianok and Kublik (2024) imply we may see these disruptions as decreased BOLD signals in fMRI data. However, our data does not demonstrate a definite pattern indicating that there were decreased BOLD signals in subjects with risk alleles. Despite this finding, fMRI BOLD signal alone wasn't used for our prediction model. Once obtaining the mean BOLD signal for each subject and adding it to our dataframe with EEG and blood test risk factors, we were able to perform logistic regression to classify risk.

When classifying AD risk using blood test, EEG, and fMRI data, our hypothesis was that the logistic regression model fit to the data would accurately predict AD risk, characterized by the presence of the genetic risk alleles APOE4 and GG in the genes APOE and PICALM. Our logistic regression model fit using 16 different factors (5 blood test, 10 EEG, and 1 fMRI) was trained on 75% of the data for a total of 30 subjects. The model predicted risk fairly accurately for the training set, with an accuracy score of 80%. It performed the best when classifying true negatives, which in the case of this study was when the actual risk was 0 and the model also predicted the risk to be 0. Testing our model on the remaining 25% of the data (10 subjects) resulted in an accuracy score of 70%. This is clearly lower than the accuracy for our training data, but is still a relatively high accuracy. Our testing data followed a similar pattern to our training data, where it performed best for true negatives when compared to other categories. The predictions for the test data also revealed that the model tended to predict no risk more often than risk, even when the subject did actually have the risk alleles. This demonstrates that the model has somewhat of a bias towards no risk, and although we included 16 different factors to classify AD risk, more factors may have been needed for the model to be more accurate in its predictions.

This leads into the limitations of this study. A big limitation of this project is the amount of subjects that the analyses were performed on. Only using 40 subjects for this study may have resulted in inaccurate modeling and limited the accuracy of the model, especially since there were only 10 subjects in our test dataset. Larger sample sizes may result in either more or less significant differences between risk subjects and non-risk subjects, and future analyses should be performed on larger cohorts of data. Another limitation of this study is that not many individual analyses were performed on the different modalities. We were unable to confirm significant differences in EEG and fMRI data between subjects with and without AD risk alleles, as the main focus of this study was to use a subset of factors from the different modalities as features of a logistic regression model. Therefore, future experiments in this area of study should analyze each modality further to truly quantify distinctions between individuals with and without AD risk alleles.

In future research, it would be interesting to use more AD risk genes than just APOE and PICALM. Adding more genes that are known to be risk factors for developing AD might aid in better classification of AD, and contribute to higher accuracy of the model. Furthermore, although we did not include psychometric data in our analyses, it would be fascinating to see whether performance on psychometric tests of memory, intelligence, mood, etc. can be associated with AD risk, and ultimately be early indicators of developing AD.

Overall, this study aimed to classify the risk of having risk alleles for Alzheimer’s disease for two genes, APOE and PICALM, using blood test, EEG, and fMRI data, and we found that the differences in these modalities between individuals with and without the risk alleles can help in predicting AD risk. Ultimately, determining various factors in multiple modalities can result in earlier detection of AD, leading to more targeted treatments and a better prognosis for individuals at risk for developing Alzheimer’s disease.
